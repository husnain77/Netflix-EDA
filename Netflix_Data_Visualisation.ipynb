```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="ML_Projects/Machine_Learning/Miscellenous/Applied Data Science-Visualisation/Netflix data/netflix_titles.csv"
df =pd.read_csv(path)
df
```

```python
# Replacments

df['country'] = df['country'].fillna(df['country'].mode()[0])


df['cast'].replace(np.nan, 'No Data',inplace  = True)
df['director'].replace(np.nan, 'No Data',inplace  = True)

# Drops

df.dropna(inplace=True)

# Drop Duplicates

df.drop_duplicates(inplace= True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Changing date
df["date_added"] = pd.to_datetime(df['date_added'])
df['month_added']=df['date_added'].dt.month
df['month_name_added']=df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year

df.head(3)
```

#Visualisaton 1

```python
import matplotlib.pyplot as plt

# Helper column for various plots
df['count'] = 1

# Many productions have several countries listed - this will skew our results , we'll grab the first one mentioned

# Lets retrieve just the first country
df['first_country'] = df['country'].apply(lambda x: x.split(",")[0])

country_data = df.groupby('first_country')['count'].sum().sort_values(ascending=False)[:10]

# Set up color map
colors = ['#4E6E81', '#4E6E81', '#4E6E81', 'lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue']

# Create plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(country_data.index, country_data, color=colors)

# Add annotations
for i, value in enumerate(country_data):
    ax.text(i, value + 50, str(value), ha='center', fontweight='bold', fontfamily='serif', fontsize=12)

# Format tick labels
ax.set_xticklabels(country_data.index, fontfamily='serif', rotation=0, fontsize=12)

# Add title and subtitle
fig.suptitle('Top 10 Titles Producing Countries on Netflix', fontsize=20, fontweight='bold', fontfamily='serif', y=1.03)
ax.set_xlabel('Country', fontsize=16, fontweight='bold', fontfamily='serif')
ax.set_ylabel('Number of Titles', fontsize=16, fontweight='bold', fontfamily='serif')

# Add grid lines
ax.grid(axis='y', linestyle='-', alpha=0.4)
ax.set_axisbelow(True)
ax.set_yticks(range(0, 3000, 500))

# Add horizontal line at 0
ax.axhline(y=0, color='black', linewidth=1.3, alpha=.7)

# Remove tick marks
ax.tick_params(axis=u'both', which=u'both', length=0)

# Remove borders
for side in ['top', 'left', 'right']:
    ax.spines[side].set_visible(False)

# Add a line at the right edge of the plot
line = plt.Line2D([1, 1], [0, 1], transform=fig.transFigure, figure=fig, color='black', lw=0.2)
fig.lines.extend([line])
fig.autofmt_xdate()

# Show the plot
plt.show()

```

# Visualisation 2: Content added by Months

```python
data_sub = df.groupby('type')['year_added'].value_counts().unstack().fillna(0).loc[['TV Show','Movie']].cumsum(axis=0).T
```

```python
month_order = ['January',
 'February',
 'March',
 'April',
 'May',
 'June',
 'July',
 'August',
 'September',
 'October',
 'November',
 'December']

df['month_name_added'] = pd.Categorical(df['month_name_added'], categories=month_order, ordered=True)
```

```python
data_sub = df.groupby('type')['month_name_added'].value_counts().unstack().fillna(0).loc[['TV Show','Movie']].cumsum(axis=0).T

plt.figure(figsize=(15, 7))
colors = ["#4E6E81", "#221f1f"]

for i, mtv in enumerate(df['type'].value_counts().index):
    mtv_rel = data_sub[mtv]
    plt.plot(mtv_rel.index, mtv_rel, color=colors[i], label=mtv, linewidth=2)

plt.axhline(y=0, color='black', linewidth=1.3, alpha=0.4)
plt.xticks(data_sub.index, rotation=0, fontfamily='serif')
plt.xlabel('Month')
plt.ylabel('Overall Total')
plt.title('Content added by month [Overall Total]')
plt.legend(['Movie', 'TV Show'])
plt.grid(False)

plt.show()
```

#Visualization 3: Ratings

```python
order = pd.DataFrame(df.groupby('rating')['count'].sum().sort_values(ascending=False).reset_index())
rating_order = list(order['rating'])
```

```python
# Create a grouped bar chart showing rating distribution for movies and TV shows
mf = df.groupby('type')['rating'].value_counts().unstack().fillna(0)[rating_order]
movie = mf.loc['Movie']
tv = - mf.loc['TV Show']
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(movie.index, movie, width=0.5, color='#4E6E81', alpha=0.8, label='Movie')
ax.bar(tv.index, tv, width=0.5, color='#221f1f', alpha=0.8, label='TV Show')
ax.set_xticklabels(mf.columns, fontfamily='serif')
ax.set_yticks([])
# Annotations
for i in tv.index:
    ax.annotate(f"{-tv[i]}", 
                   xy=(i, tv[i] - 60),
                   va = 'center', ha='center',fontweight='light', fontfamily='serif',
                   color='#4a4a4a')   

for i in movie.index:
    ax.annotate(f"{movie[i]}", 
                   xy=(i, movie[i] + 60),
                   va = 'center', ha='center',fontweight='light', fontfamily='serif',
                   color='#4a4a4a')
    
ax.legend(loc='upper right', frameon=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Rating Distribution by Film & TV Show', fontweight='bold', fontfamily='serif')

plt.show()
```
